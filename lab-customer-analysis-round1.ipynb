```python
#STANDARDISE HEADER
#CREATE A LIST OF DF NAMES
df_list = [df1, df2, df3]
col1, col2, col3 = [],[],[]

#FD1
for col in df1.columns:
    col1.append(col.lower())
    
#DF2
for col in df2.columns:
    col2.append(col.lower())
    
#DF3
for col in df3.columns:
    col3.append(col.lower())
display(col1)

        
```

```python
df1.columns = col1
df2.columns = col2
df3.columns = col3
display(df1.columns)
display(df2.columns)
display(df3.columns)
```

```python
#CHECK IF COLUMNS ARE IDENICAL
sorted(col1) == sorted(col2) == sorted(col3)
```

```python
if len(col1) == len(col1) == len(col1):
    display('LISTS EQUAL')
```

```python
#USING col3 AS THE STANDARD
#GET VALUES THAT ARE IN COL1 & COL2 BUT NOT IN COL3
for item in col1:
    if item not in col3:
        print(item)
```
